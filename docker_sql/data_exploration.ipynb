## NY CAB data exploration

```python
import pandas as pd
from urllib import request
```

```python
URL = "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_{}.parquet"
      https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2022-1.parquet
```

```python
date_month = '2022-01'
response = request.urlretrieve(URL.format(date_month), f"yellow_tripdata_{date_month}.parquet")
response
```

```python
data = pd.read_parquet(f"yellow_tripdata_{date_month}.parquet")
data.info()
```

```python
data.head()
```

```python
taxi_zones = "https://d37ci6vzurychx.cloudfront.net/misc/taxi+_zone_lookup.csv"
response = request.urlretrieve(taxi_zones, "taxi_zones.csv")
response
```

```python
df = pd.read_csv("taxi_zones.csv")
df.head()
```

```python
borough =df.set_index('LocationID').to_dict('index')
```

```python
data['PULocation_Borough'] = data.apply(lambda x: borough[x['PULocationID']]['Borough'],axis=1) 
data['PULocation_Zone'] = data.apply(lambda x: borough[x['PULocationID']]['Zone'],axis=1) 
```
